## Apriori Algorithm

```python
pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_items=apriori(book,min_support=0.1,use_colnames=True)
frequent_items
```

```python
print(len(frequent_items))
```

```python
frequent_items.sort_values("support",ascending=False)
```

```python
rules=association_rules(frequent_items,metric="lift",min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules_final=rules[rules.lift>1]
rules_final
```

```python
# Changing minimum values of minimum length in apriori 
frequent_items1=apriori(book,min_support=0.1,use_colnames=True)
frequent_items1
```

```python
rules1=association_rules(frequent_items,metric="lift",min_threshold=0.9)
rules1
```

```python
rules_final
```

## Visualisation

```python
rules_final['antecedents_']=rules_final['antecedents'].apply(lambda a: ','.join(list(a)))
rules_final['consequents_']=rules_final['consequents'].apply(lambda a: ','.join(list(a)))

# Transform the DataFrame of rules into a matrix using the confidence metric

pivot=rules_final[rules_final['lift']>1].pivot(index='antecedents_',columns='consequents_',values='confidence')

# Generate a heatmap with annotations 

plt.figure(figsize=(10,5))
sns.heatmap(pivot,annot=True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

## Apriori Algorithm

```python
frequent_movies=apriori(movie1,min_support=0.1,use_colnames=True)
frequent_movies
```

```python
rules_movie=association_rules(frequent_movies,metric="lift", min_threshold=0.7)
rules_movie
```
